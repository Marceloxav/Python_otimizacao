```python
import numpy as np
from scipy.optimize import fmin
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x) - 3 * np.exp(-(x - 0.2) ** 2)

# determina minimo de f(x)
minimum1 = fmin(f, 1.0)
print(f"Inicia a pesquisa com x=1., mínimo é {minimum1[0]:.6f} com valor {f(minimum1)[0]:.6f}")

minimum2 = fmin(f, 2.0)
print(f"Inicia a pesquisa com x=2., mínimo é {minimum2[0]:.6f} com valor {f(minimum2)[0]:.6f}")

# plota função
x = np.arange(-10, 10, 0.1)  # arange agora vem do numpy
y = f(x)
plt.plot(x, y, 'b-', linewidth=2, label='$\\cos(x)-3e^{-(x-0.2)^2}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axis([-5, 5, -2.5, 0.5])

# adiciona os mínimos e pontos iniciais
plt.plot(minimum1, f(minimum1), 'vr', markersize=10, label=f'mínimo 1 em x={minimum1[0]:.3f}')
plt.plot(1.0, f(1.0), 'or', markersize=8, label='start 1 (x=1.0)')
plt.plot(minimum2, f(minimum2), 'vg', markersize=10, label=f'mínimo 2 em x={minimum2[0]:.3f}')
plt.plot(2.0, f(2.0), 'og', markersize=8, label='start 2 (x=2.0)')

plt.legend(loc='lower left')
plt.title('Minimização local com fmin')
plt.show()
```
